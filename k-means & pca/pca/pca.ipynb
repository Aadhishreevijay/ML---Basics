```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('diabetes.csv')
x=data.iloc[:,:-1].values
```

```python
mean=np.mean(x,axis=0)
```

```python
centered_data=x-mean
```

```python
cov=np.cov(centered_data,rowvar=False)
```

```python
eig_val,eig_vec=np.linalg.eig(cov)
```

```python
s_i=np.argsort(eig_val)[::-1]
eig_val_sor=eig_val[s_i]
eig_vec_sor=eig_vec[:,s_i]
```

```python
k=2
f_v=eig_vec_sor[:,:k]
```

```python
new_data = centered_data.dot(f_v)
```

```python
plt.subplot(1,2,1)
plt.scatter(data.iloc[:,0],data.iloc[:,1],alpha=0.5)
plt.title('original')
plt.xlabel('feature 1')
plt.ylabel('feature 2')

plt.subplot(1,2,2)
plt.scatter(new_data[:,0],new_data[:,1],alpha=0.5)
plt.title('new')
plt.xlabel('feature 1')
plt.ylabel('feature 2')

plt.show()
```
