```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('titanic.csv')
numerical_data=data.select_dtypes(include=[np.number]).dropna().values
x=numerical_data[:,:-1]
```

```python
mean=np.mean(x,axis=0)
```

```python
centered_data=x-mean
```

```python
cov=np.cov(centered_data,rowvar=False)
```

```python
eig_values,eig_vec=np.linalg.eig(cov)
```

```python
sorted=np.argsort(eig_values)[::-1]
eig_val_sort=eig_values[sorted]
eig_vec_sort=eig_vec[:,sorted]
```

```python
k=2
f_v=eig_vec_sort[:,:k]
```

```python
new_data=centered_data.dot(f_v)
```

```python
plt.subplot(1,2,1)
plt.scatter(numerical_data[:,0],numerical_data[:,1],alpha=0.5)
plt.title('og')
plt.xlabel('f1')
plt.ylabel('f2')
plt.subplot(1,2,2)
plt.scatter(new_data[:,0],new_data[:,1],alpha=0.5)
plt.title('new')
plt.xlabel('f1')
plt.ylabel('f2')
plt.show()
```
