```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('titanic.csv')
k=5
numerical_data = data.select_dtypes(include=[np.number]).dropna().values
```

```python
def e_d(p1,p2):
    return np.sqrt(np.sum((p1-p2)**2))
```

```python
def ini_c(data,k):
    return data[np.random.choice(range(len(data)),k,replace=False)]
```

```python
def assign(data,centroids):
    distance=np.array([[e_d(point,centroid) for centroid in centroids] for point in data])
    return np.argmin(distance,axis=1)
```

```python
def update(data,labels,k):
    new_centroid=np.zeros((k,data.shape[1]))
    for i in range(k):
        c_p=data[labels==i]
        if len(c_p)>0:
            new_centroid[i]=np.mean(c_p,axis=0)
    return new_centroid
```

```python
def visu(data,k,itr=10):
    centroids=ini_c(data,k)
    for i in range(itr):
        labels=assign(data,centroids)
        new_centroid=update(data,labels,k)
        plt.figure(figsize=(6,6))
        plt.scatter(data[:,0],data[:,1],c=labels,cmap='viridis',edgecolor='k',s=25)
        plt.scatter(new_centroid[:,0],new_centroid[:,1],c='red',marker='x',s=200,label=f'centroid {itr+1}')
        plt.title(f'k cluster {itr+1}')
        plt.xlabel('feature 1')
        plt.ylabel('feature 2')
        plt.show
        if np.all(centroids==new_centroid):
            break
        centroids=new_centroid
    return labels,centroids
```

```python
visu(numerical_data,k)
```
